### Importing

```python
import pandas as pd
```

### Importing the CSV Folders of r/Bipolar and r/Schizophrenia

```python
# Assuming you have two dataframes df_bipolar and df_schizophrenia
df_bipolar = pd.read_csv('../data/bipolar.csv')
df_schizophrenia = pd.read_csv('../data/schizophrenia.csv')
```

#### Dropping the repeat rows from the two csv files and exporting it

```python
#For each of the dataframes, add a column called 'label' with the value 0 for bipolar and 1 for schizophrenia
df_bipolar['label'] = 0
df_schizophrenia['label'] = 1
```

```python
# Concatenate the two dataframes
df = pd.concat([df_bipolar, df_schizophrenia])

# Print the number of duplicates
print("Number of duplicate rows: ", df.duplicated().sum())
```

```python
#Find the columns of the new df
print("Columns of the new df: ", df.columns)
```

```python
#Find number of duplicated ids
print("Number of duplicated ids: ", df['id'].duplicated().sum())
```

```python
#Find total number of rows
print("Total number of rows: ", df.shape[0])
```

```python
#Drop rows with duplicated ids
df = df.drop_duplicates(subset=['id'])

# Reset index after dropping duplicates
df = df.reset_index(drop=True)

# Print the number of rows now
print("Total number of rows after dropping duplicates: ", df.shape[0])
```

```python
#display entire length of column instead of conjunguated view
pd.set_option('display.max_colwidth', -1)

#See the column of text and score for the label 1, in the column of text
df[df['label'] == 1]['text']

df[df['label'] == 0]['text']
```

```python
#For the columns text with the label of 0, find counts of all the words the exists in these columns and then return a dataframe with all the words and their corresponding counts
df[df['label'] == 0]['text'].str.split(expand=True).stack().value_counts().head(50)

```

```python
#Find the same counts for the label 1
df[df['label'] == 1]['text'].str.split(expand=True).stack().value_counts().head(50)
```

```python
#Make a new list showing the proportion of each word for both text columns
df[df['label'] == 0]['text'].str.split(expand=True).stack().value_counts(normalize=True).head(50)

```

```python
# Save the dataframe to a new csv file
df.to_csv('../data/combined.csv', index=False)
```
